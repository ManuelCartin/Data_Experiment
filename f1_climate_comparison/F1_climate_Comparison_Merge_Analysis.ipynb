# Spa Francorchamp Circuit
The selection of Durbuy, Belgium, as one of the reference points for this study is no coincidence. Although it is not directly located at the coordinates of the legendary Spa-Francorchamps circuit, its geographical proximity makes it the most representative weather station available to capture the region's unique climatic conditions.

The Spa-Francorchamps circuit, located in the Belgian Ardennes, is known worldwide as "the temple of speed" and is one of the most challenging and beloved circuits for drivers in Formula 1 and other motorsports categories. Its winding 7.004 km layout, which runs through dense forests and hills, is famous for its iconic combination of high-speed corners such as Eau Rouge and Raidillon.

However, what truly sets Spa apart is the unpredictability of its weather. Often, one section of the circuit can be completely dry while another, just a few kilometers away, suffers from torrential rain. This phenomenon has led to some of the most exciting and chaotic races in the history of motorsport.

By using data from Durbuy, our analysis seeks to capture the essence of this unstable climate: the rapid temperature fluctuations, high humidity levels, and constant threat of precipitation that are characteristic of this region of Belgium. In this way, the data don't just represent numbers, but tell the story of a climate that is, in itself, a determining factor in the drama and strategy of motor racing.

# Singapour
The Marina Bay Street Circuit in Singapore is famous for being Formula 1's first night Grand Prix. Beyond its visual spectacularity, the weather conditions, experienced directly in the city, are the true challenge that elevates this race to a level of its own.

Singapore has an equatorial climate, which results in high temperatures and extremely high humidity year-round, with no major seasonal variations. This creates a sticky, dense atmosphere, which makes it difficult for the engines to cool down and tests the endurance of the drivers, who must endure the overwhelming heat inside their cockpits. Humidity also affects visibility and can fog up helmet visors, even with the cars' air conditioning.

Unlike other night races, where temperatures drop significantly, in Singapore the drop is minimal, maintaining the physical and technical challenge throughout the two-hour race. Furthermore, the risk of torrential rain is a constant. These rains, common in the region, can turn the circuit into a slippery and dangerous surface in a matter of minutes, forcing teams to be constantly ready to change strategy.

Analyzing Singapore's climate data allows us to immerse ourselves in this extreme environment, where heat and humidity are not just statistical variables, but living elements that interact with every lap of the car and every decision made by the teams.

```python
climate_singapour = pd.read_csv('/content/sample_data/export-singapour0.csv', skiprows=3)
display(climate_singapour.head())
```

```python
climate_singapour.columns = [
  col if col == 'DATE' else f"{col}_singapour"
  for col in climate_singapour.columns
]
display(climate_singapour.head())
```

# Merge

```python
from functools import reduce
dfs = [climate_durbuy, climate_manama, climate_miami, climate_singapour]
climate_global = reduce(lambda left, right: pd.merge(left, right, on='DATE', how='inner'), dfs)
```

```python
display(climate_global.head())
```

```python
print(climate_global.columns.tolist())
```

```python
climate_global.to_csv('climate_global_f1_circuit.csv', index=False)
```

# Data vis

```python
climate_global.plot(
    x='DATE',
    y=[
        'MAX_TEMPERATURE_C_durbuy',
        'MAX_TEMPERATURE_C_manama',
        'MAX_TEMPERATURE_C_miami',
        'MAX_TEMPERATURE_C_singapour'
    ],
    figsize=(14,6),
    title='Comparaci√≥n de temperaturas m√°ximas por ciudad'
)
plt.show()
```

```python
climate_global['MONTH'] = pd.to_datetime(climate_global['DATE']).dt.month
monthly_avg = climate_global.groupby('MONTH')[
    [col for col in climate_global.columns if 'MAX_TEMPERATURE_C' in col]
]. mean()
sns.heatmap(monthly_avg.T, cmap='coolwarm', annot=True)
plt.title('Temperaturas m√°ximas por mes')
plt.show()
```

```python
monthly_humidity = climate_global.groupby('MONTH')[
    [col for col in climate_global.columns if 'HUMIDITY_MAX_PERCENT' in col]
  ].mean()
sns.heatmap(monthly_humidity.T, cmap='YlGnBu', annot=True)
plt.title('Humedad maxima por mes')
plt.show()
```

```python
for city in ['durbuy', 'manama', 'miami', 'singapour']:
  climate_global[f'Amplitud_{city}'] =climate_global[f'MAX_TEMPERATURE_C_{city}'] - climate_global[f'MIN_TEMPERATURE_C_{city}']
amplitud_avg = climate_global.groupby('MONTH')[[f'Amplitud_{city}' for city in ['durbuy', 'manama', 'miami', 'singapour']]].mean()
sns.heatmap(amplitud_avg.T, cmap='magma', annot=True)
plt.title('Amplitud de temperaturas por mes')
plt.show()
```
